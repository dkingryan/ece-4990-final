```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt

# Load a COCO-pretrained YOLO26n model
model = YOLO("yolo26m.pt") # change to the yolo model you want to use
# print(model.names)

# Train the model on the COCO8 example dataset for 100 epochs
results = model.train(data="data.yaml", epochs=200, imgsz=640)

# Run inference with the YOLO26n model on the 'bus.jpg' image
results = model("human.jpg")

# # first image result
r = results[0]

# # draw boxes + labels + confidence onto the image
annotated = r.plot()

plt.figure(figsize=(10, 10))
plt.imshow(annotated)
plt.axis("off")
plt.show()

# Validate the model
metrics = model.val(data='coco.yaml', classes=[0])  # no arguments needed, dataset and settings remembered
metrics.box.map  # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps  # a list containing mAP50-95 for each category
metrics.box.image_metrics  # per-image metrics dictionary with precision, recall, F1, TP, FP, and FN

print(f"mAP50-95: {metrics.results_dict}")
```
